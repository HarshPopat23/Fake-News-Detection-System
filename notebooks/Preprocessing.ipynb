```python
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = pd.read_csv("../data/True.csv")
df_fake = pd.read_csv("../data/Fake.csv")

df["label"] = 1
df_fake["label"] = 0

df = pd.concat([df, df_fake]).sample(frac=1).reset_index(drop=True)
```

```python
def remove_punc(txt):
    return txt.translate(str.maketrans('', '', string.punctuation))

def remove_num(txt):
    return ''.join([i for i in txt if not i.isdigit()])

def remove_emoji(txt):
    return ''.join([i for i in txt if i.isascii()])
```

```python
df["title"] = df["title"].astype(str).apply(remove_num).apply(remove_punc).apply(remove_emoji)
df["text"] = df["text"].astype(str).apply(remove_num).apply(remove_punc).apply(remove_emoji)
```

```python
nltk.download("punkt")
nltk.download("stopwords")

stop_words = set(stopwords.words("english"))

def remove_stopwords(txt):
    words = word_tokenize(txt)
    return " ".join([w for w in words if w not in stop_words])

df["title"] = df["title"].apply(remove_stopwords)
df["text"] = df["text"].apply(remove_stopwords)
```

```python
df["content"] = df["title"] + " " + df["text"]
df[["content", "label"]].head()
```
